```python
from scipy import stats
import pandas as pd
import numpy as np
```

## Analyze

```python
df = pd.read_csv('clean_data.csv')
df
```

```python
num_rec = df.shape[0]
```

### App used analyze

```python
app_used = df['apps used']
app_used
```

```python
app_list = [[item.strip() for item in elem.split(",")] for elem in app_used]
flatten_list = np.array([item for sublist in app_list for item in sublist])
unique_elements, counts = np.unique(flatten_list, return_counts=True)
frequency_dict = dict(zip(unique_elements, counts/num_rec*100))
count_dict = dict(zip(unique_elements, counts))

sorted_frequency_dict = dict(sorted(frequency_dict.items(), key=lambda item: item[1], reverse=True))
sorted_frequency_dict
rounded_frequency_dict = {key: round(value, 2) for key, value in sorted_frequency_dict.items()}
rounded_frequency_dict
```

```python
count_dict = dict(sorted(count_dict.items(), key=lambda item: item[1], reverse=True))
count_dict
```

### PEOU vs Domain Proficiancy PU analyze

```python
rec_list = [[item.strip() for item in elem.split(",")] for elem in reccommendation]
flatten_rec_list = np.array([item for sublist in rec_list for item in sublist])
rec_unique_elements, rec_counts = np.unique(flatten_rec_list, return_counts=True)
rec_frequency_dict = dict(zip(rec_unique_elements, rec_counts/num_rec*100))
rec_count_dict = dict(zip(rec_unique_elements, rec_counts))

rec_sorted_frequency_dict = dict(sorted(rec_frequency_dict.items(), key=lambda item: item[1], reverse=True))
rec_sorted_frequency_dict
# rounded_frequency_dict = {key: round(value, 2) for key, value in sorted_frequency_dict.items()}
# rounded_frequency_dict
```

```python
import statsmodels.api as sm

X = sm.add_constant(df['ease of use'])
model_usefulness  = sm.OLS(df['perceived usefulness'], X).fit()
model_domain = sm.OLS(df['domain proficiency'], X).fit()

print(model_usefulness.summary())
print(model_domain.summary())
```

```python
X = sm.add_constant(df['domain proficiency'])
model_dp_eou = sm.OLS(df['ease of use'], X).fit()

print(model_dp_eou.summary())
```

```python
X = sm.add_constant(df['ease of use'])
model_peou_i  = sm.OLS(df['intention'], X).fit()

print(model_peou_i.summary())
```
