```python
## Hàm viết tắt
def abbreviation(sentence):

  if ' ' not in sentence:
    return sentence

  words = sentence.split()
  first_letters_list = [word[0] for word in words]
  first_letters_string = ''.join(first_letters_list)
  
  return first_letters_string
```

```python
abbreviation("tài chính doanh nghiệp")
```

```python
new_df['Major abbreviation'] = new_df['raw major'].apply(abbreviation)
```

```python
def recommendation_classify(str):
  arr = [item.strip() for item in str.split(',')]
  res = []
  ## Handle special values
  social_media_values = ['Biết từ Youtube -> Muốn thử xem thế nào', 'Của trình duyệt', 'Gợi ý từ các nền tảng mạng xã hội',
                         'Các thông tin từ các phương tiện truyền thông (báo chí', '...)' ]
  friends_reccommend_values = ['Đề xuất của bạn bè']
  for elem in arr:
    if elem in social_media_values:
      res.append('social media')
    elif elem in friends_reccommend_values:
      res.append('friends recommendation')
    else:
      res.append('independent discovery')
    # print(elem, res)

  return ', '.join(list(set(res)))
```
